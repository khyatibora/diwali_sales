# Exploratory Data Analysis

# Gender


```python
ax = sns.countplot(x = 'category_by_Gender', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['category_by_Gender'], as_index=False)['OriginalPrice (in Rs)'].sum().sort_values(by=['OriginalPrice (in Rs)'],ascending = False)
sns.barplot(x='category_by_Gender', y = 'OriginalPrice (in Rs)', data = sales_gen)
```

From above graphs we can see that most of the buyers are Men and even the purchasing power of the men are greater than women

# Individual_category

```python
sns.set(rc={'figure.figsize':(15,10)})

top_categories = df['Individual_category'].value_counts().head(8).index


filtered_df = df[df['Individual_category'].isin(top_categories)]


ax = sns.countplot(data=filtered_df, x='Individual_category')


for bars in ax.containers:
    ax.bar_label(bars)

# Show the plot
plt.show()
```

```python
sales_state=df.groupby(['Individual_category'],as_index=False)['OriginalPrice (in Rs)'].sum().sort_values(by='OriginalPrice (in Rs)',ascending = False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_state,x='Individual_category',y='OriginalPrice (in Rs)',hue='Individual_category')
```

From above graphs we can see the most of the sold products are from Kurta-sets,jeans and kurtas

# Average Ratings

```python
ax = sns.countplot(x = 'Ratings', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=df.groupby(['category_by_Gender'],as_index=False)['Ratings'].sum().sort_values(by='Ratings',ascending = False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_state,x='category_by_Gender',y='Ratings',hue='category_by_Gender')
```

From above graphs we can see that the ratings in men-based products are higher than women-based products

#  Discount Offer

```python
df['DiscountOffer'] = pd.to_numeric(df['DiscountOffer'], errors='coerce')


sales_state = df.groupby('Category')['DiscountOffer'].sum().nlargest(5).reset_index()


sns.set(rc={'figure.figsize':(12, 6)})  

# Create the barplot
sns.barplot(data=sales_state, x='Category', y='DiscountOffer')

# Adding labels and title
plt.xlabel('Category')
plt.ylabel('Discount Offer Sum')
plt.title('Top 5 Categories by Discount Offer')

# Show the plot
plt.show()
```

# Discount Price by Brand

```python

top_10_prices = df['DiscountPrice (in Rs)'].value_counts().nlargest(10).index


df_top_10 = df[df['DiscountPrice (in Rs)'].isin(top_10_prices)

ax = sns.countplot(x='DiscountPrice (in Rs)', data=df_top_10)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=df.groupby(['BrandName'],as_index=False)['DiscountPrice (in Rs)'].sum().sort_values(by='DiscountPrice (in Rs)',ascending = False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_state,x='BrandName',y='DiscountPrice (in Rs)',hue='BrandName')
```

 Puma, here&now brands offer the largest discounts on average.



# Average Ratings by Category

```python
top_10_prices = df['Category'].value_counts().nlargest(10).index


df_top_10 = df[df['Category'].isin(top_10_prices)]

ax = sns.countplot(x='Category', data=df_top_10)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=df.groupby(['Category'],as_index=False)['Ratings'].sum().sort_values(by='Ratings',ascending = False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_state,x='Category',y='Ratings',hue='Category')
```

From the above graphs Indian wears has the highest rating by category.

# Total Number of Reviews by Brand

```python
top_10_prices = df['Reviews'].value_counts().nlargest(10).index


df_top_10 = df[df['Reviews'].isin(top_10_prices)]

ax = sns.countplot(x='Reviews', data=df_top_10)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=df.groupby(['BrandName'],as_index=False)['Reviews'].sum().sort_values(by='Reviews',ascending = False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_state,x='BrandName',y='Reviews',hue='BrandName')
```

 From above graphs Roadster, HERE&NOW brands have the most reviews.
